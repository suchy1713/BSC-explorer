```python
import json
import os
import pandas as pd
input_directory = 'input/'
only_passes = True

dfs = []
for filename in os.listdir(input_directory):
    if filename.endswith(".json"):
        with open(input_directory + filename, 'r') as f:
            print(filename)
            data = json.load(f)

    events = pd.DataFrame(data['events'])
    events['home_team_name'] = data['home_name']
    events['away_team_name'] = data['away_name']
    events['date'] = data['date']
    events['competition'] = data['competition']

    if only_passes:
        events = events.loc[events['type'] == 'Pass']

    dfs.append(events)

final_df = pd.concat(dfs)
final_df.to_csv('passes.csv', index=False)
```
