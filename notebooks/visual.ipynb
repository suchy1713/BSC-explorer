```python
import pandas as pd
import matplotlib.pyplot as plt
from field import draw_field
from plot import set_style

df = pd.read_csv('../model/3_merge_features_and_positions/data/data.csv')
df = df[['x', 'y', 'position']]
print(df.shape)
df.head()
```

```python
color_pos = {
    'CB': 'olivedrab',
    'LB': 'mediumblue',
    'RB': 'red',
    'CM': 'lightskyblue',
    'LW': 'fuchsia',
    'RW': 'darkorange',
    'ST': 'limegreen',
}

fig = plt.figure(figsize=(30, 20))
ax = plt.axes([0.05, 0.05, 0.9, 0.9])
set_style()
draw_field(ax, correct_aspect=True, arrow=False)

ax.scatter(df['x'], df['y'], zorder=1000, color=df['position'].map(color_pos), alpha=0.8, linewidth=0.5, edgecolor='#cdcdcd', s=100)

fig.savefig('positions.png', dpi=300)
```
